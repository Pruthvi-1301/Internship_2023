```python
#Morphological Transformations

fig = plt.figure()
kernel = np.ones((5,5),np.uint8)

binary = cv.imread("Binary.jpg")
ax = fig.add_subplot(3, 3, 1)
ax.set_title("original")
ax.imshow(binary)

erosion = cv.erode(binary, kernel,iterations = 1)
ax1 = fig.add_subplot(3, 3, 2)
ax1.set_title("Erosion")
ax1.imshow(erosion)

dilation = cv.dilate(binary, kernel,iterations = 1)
ax2 = fig.add_subplot(3, 3, 3)
ax2.set_title("Dilation")
ax2.imshow(dilation)

opening = cv.morphologyEx(binary, cv.MORPH_OPEN, kernel)
ax3 = fig.add_subplot(3, 3, 4)
ax3.set_title("Opening")
ax3.imshow(erosion)

closing = cv.morphologyEx(binary, cv.MORPH_CLOSE, kernel)
ax3 = fig.add_subplot(3, 3, 5)
ax3.set_title("Closing")
ax3.imshow(closing)

gradient = cv.morphologyEx(binary, cv.MORPH_GRADIENT, kernel)
ax4 = fig.add_subplot(3, 3, 6)
ax4.set_title("Gradient")
ax4.imshow(gradient)

tophat = cv.morphologyEx(binary, cv.MORPH_TOPHAT, kernel)
ax5 = fig.add_subplot(3, 3, 7)
ax5.set_title("TopHat")
ax5.imshow(tophat)

blackhat = cv.morphologyEx(binary, cv.MORPH_BLACKHAT, kernel)
ax6 = fig.add_subplot(3, 3, 8)
ax6.set_title("BlackHat")
ax6.imshow(blackhat)
```
