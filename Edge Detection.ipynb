```python
#Canny Edge Detection
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('Edge_Detection.jpg')
gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
blur = cv.GaussianBlur(img, (3,3), 0)

edges = cv.Canny(image=blur, threshold1=100, threshold2=200)

fig = plt.figure()
ax1 = fig.add_subplot(3, 3, 1)
ax1.set_title("Regular Image")
ax1.imshow(img)

ax2 = fig.add_subplot(3, 3, 2)
ax2.set_title("Edge Detected")
ax2.imshow(edges)
```

```python
#Sobel Edge Detection
img = cv.imread('Edge_Detection.jpg')
gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
blur = cv.GaussianBlur(img, (3,3), 0)

sobelx = cv.Sobel(src=blur, ddepth=cv.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv.Sobel(src=blur, ddepth=cv.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv.Sobel(src=blur, ddepth=cv.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection

# Display Sobel Edge Detection Images
fig = plt.figure()
ax1 = fig.add_subplot(3, 3, 1)
ax1.set_title("Sobel X")
ax1.imshow(sobelx)

ax2 = fig.add_subplot(3, 3, 2)
ax2.set_title("Sobel Y")
ax2.imshow(sobely)

ax3 = fig.add_subplot(3, 3, 3)
ax3.set_title("Combined")
ax3.imshow(sobelxy)
```
